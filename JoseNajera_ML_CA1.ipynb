```python
# Calcio vs Type of Glass

sns.catplot(y = "Calcium", x = "Glass_type", data = dataset_1.sort_values("Calcium", ascending = False), kind="boxen", height = 6, aspect = 4)
plt.title('Calcium vs Glass_type')
plt.show()
```

```python
# Magnesium vs Type of Glass

sns.catplot(y = "Magnesium", x = "Glass_type", data = dataset_1.sort_values("Magnesium", ascending = False), kind="boxen", height = 6, aspect = 4)
plt.title('Magnesium (Mg) vs Glass_type')
plt.show()
```
